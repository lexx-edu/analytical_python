4. Создать новый признак delta_renovated, который будет содержать разницу в годах между годом реновацией дома и годом постройки дома. Если реновации дома не было, то в новом признаке поставьте 0

```python
df['delta_renovated'] = df['yr_renovated'] - df['yr_built']
df.loc[df['delta_renovated']<0, 'delta_renovated'] = 0

df.head()
```

5. Создайте признаки года продажи, месяца продажи

```python
df['date'] = pd.to_datetime(df['date'], dayfirst=False).dt.date

df['year'] = df['date'].apply(lambda x: x.year)
df['month'] = df['date'].apply(lambda x: x.month)

df.head()
```

6. Удалите признаки date, zipcode, lat, long

```python
df.drop(['zipcode', 'lat', 'long'], axis=1, inplace=True)

df.head()
```

## Задание 03

1. Найдите среднюю стоимость домов в зависимости от количества спален. Отсортируйте от меньшей стоимости к большей

```python
df.groupby('bedrooms', as_index=False).agg({'price': 'mean'}).sort_values('price')
```

2. Найдите минимальную, среднюю и максимальную стоимости домов в зависимости от состояния дома

```python
df[['condition', 'price']].groupby('condition', as_index=False).agg(['min', 'mean', 'max'])
```

3. Постройте таблицу с подсчетом количества домов в данных в зависимости от вида на набережную и оценкой вида

```python
df.pivot_table(index='waterfront', columns='view', values='id', aggfunc='count', fill_value=0)
```

4. Каких домов в зависимости от этажности и количества спален больше?

```python
tmp = df.pivot_table(index='floors', columns='bedrooms', values='id', aggfunc='count', fill_value=0)

tmp
```

```python
col_max = tmp.max().sort_values(ascending=False).index[0]
row_max = tmp.max(axis=1).sort_values(ascending=False).index[0]

tmp.loc[tmp.index == row_max, tmp.columns == col_max]
```
